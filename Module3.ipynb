```python
#Read numbers from a file and write even and odd numbers to separate files
f = open("D:/222/even_odd.txt", "r")
string = f.read()
x = string.split()
even = []
odd = []
for i in range(0, len(x)):
    x[i] = int(x[i])
for a in x:
    if a%2 == 0:
        b = str(a)
        f = open("D:/222/even.txt", "a")
        f.write(b)
        f.write(" ")
        f.close()
    else:
        b = str(a)
        f = open("D:/222/odd.txt", "a")
        f.write(b)
        f.write(" ")
        f.close()

f = open("D:/222/even.txt", "r")
print(f.read())
f = open("D:/222/odd.txt", "r")
print(f.read())
```
